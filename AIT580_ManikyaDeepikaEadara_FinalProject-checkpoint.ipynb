```python
import numpy as np
import pandas as pd
data = pd.read_csv("https://data.ny.gov/api/views/5i6e-asw6/rows.csv")
data.head()
```

```python
# Remove any leading or trailing whitespace from all column names
data.columns = data.columns.str.strip()
data.head()
```

```python
# Rename the "MT CO2e AR5 20 yr" and "MT CO2e AR4 100 yr" columns to more manageable names
data.rename(columns={"MT CO2e AR5 20 yr": "co2e_20yr", "MT CO2e AR4 100 yr": "co2e_100yr"}, inplace=True)
data.head()
```

```python
# Convert the "Gross", "Net", "Conventional Accounting", "MT CO2e AR5 20 yr", and "MT CO2e AR4 100 yr" columns to numeric
data[["Gross", "Net", "Conventional Accounting", "co2e_20yr", "co2e_100yr"]] = data[["Gross", "Net", "Conventional Accounting",
 "co2e_20yr", "co2e_100yr"]].apply(pd.to_numeric)


```

```python
data.replace({"Yes": True, "No": False}, inplace=True)
data.head()
```

```python
# Drop any rows with missing values
data.dropna(inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#univariate Analysis
# Box plot of greenhouse gas emissions by sub-category:
import plotly.express as px

fig = px.box(data, x='Sub-Category 1', y='co2e_100yr',
             title='Greenhouse Gas Emissions by Sub-Category in New York State')
fig.update_xaxes(title='Sub-Category')
fig.update_yaxes(title='Greenhouse Gas Emissions (CO2e_100 yr)')

fig.show()
```

```python
# Aggregate emissions by year and economic sector
emissions_by_sector_year = data.groupby(['Sector', 'Year'])['co2e_100yr'].mean().reset_index()

# Create line chart
plt.figure(figsize=(10, 7))
sns.lineplot(x="Year", y="co2e_100yr", hue="Sector", data=emissions_by_sector_year)
plt.title('Greenhouse Gas Emissions by Economic Sector and Year in New York State')
plt.xlabel('Year')
plt.ylabel('Greenhouse Gas Emissions (MT CO2e AR4 100 yr)')
plt.show()
```

```python
# Linear regression plot of MT CO2e AR4 100 yr by Year and Sector
sns.lmplot(x="Year", y="co2e_100yr", hue="Sector", data=data)
plt.xlabel("Year")
plt.ylabel("CO2e_100 yr (MT)")
plt.title("CO2e_100 yr by Sector over Time")
plt.show()
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')


# Tokenize the 'Sector' column
tokens = []
for sector in data['Sector']:
    tokens.extend(word_tokenize(sector))

# Remove stopwords and lowercase tokens
stop_words = set(stopwords.words('english'))
filtered_tokens = [w.lower() for w in tokens if w.lower() not in stop_words]



```

```python
# Count the frequency of each term
freq_dist = nltk.FreqDist(filtered_tokens)


# Display the 10 most common terms
print(freq_dist.most_common(10))
```

```python
#Research Question 3

import requests
import pandas as pd
import io

api_key = "7e6a7d6ddf28f4a070c3650927d55e12"
series_id = "NYPOP"
start_year = 1990
end_year = 2019

# Prepare the URL
url = f"https://fred.stlouisfed.org/graph/fredgraph.csv?mode=fred&id={series_id}&api_key={api_key}&file_type=csv"

# Send the API request
response = requests.get(url)

# Convert the response content to a pandas DataFrame
population_data = pd.read_csv(io.StringIO(response.text), parse_dates=["DATE"])

# Filter the data for the desired years
population_data = population_data[(population_data["DATE"].dt.year >= start_year) & (population_data["DATE"].dt.year <= end_year)]

# Rename columns for better readability
population_data.rename(columns={"DATE": "Year", series_id: "Population"}, inplace=True)
population_data["Year"] = population_data["Year"].dt.year

print(population_data)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from scipy import stats


# Sum the greenhouse gas emissions by year
annual_ghg_emissions = data.groupby("Year")["co2e_20yr"].sum().reset_index()
annual_ghg_emissions.rename(columns={"co2e_20yr": "GHG_Emissions"}, inplace=True)

# Merge the annual_ghg_emissions and population_data DataFrames on the 'Year' column
merged_data = annual_ghg_emissions.merge(population_data, on="Year")

# Calculate population density (people per square mile)
ny_state_area_sq_miles = 54555
merged_data["Population Density"] = merged_data["Population"] / ny_state_area_sq_miles

# Calculate the correlation between population density and GHG emissions
correlation = np.corrcoef(merged_data["Population Density"], merged_data["GHG_Emissions"])[0, 1]
print(f"Correlation between population density and GHG emissions: {correlation:.2f}")

# Calculate trendline using numpy.polyfit
x = merged_data["Population Density"]
y = merged_data["GHG_Emissions"]
slope, intercept = np.polyfit(x, y, 1)
merged_data["Trendline_Slope"] = slope
merged_data["Trendline_Intercept"] = intercept

# Create scatter plot with trendline
fig = px.scatter(merged_data, x="Population Density", y="GHG_Emissions",
                 title="Population Density vs. GHG Emissions in New York State",
                 labels={"Population Density": "Population Density (people per square mile)",
                         "GHG_Emissions": "Greenhouse Gas Emissions (co2e_20yr)"})
fig.add_trace(go.Scatter(x=merged_data["Population Density"],
                         y=merged_data["Population Density"] * merged_data["Trendline_Slope"] + merged_data["Trendline_Intercept"],
                         mode="lines",
                         line=dict(color="red"),
                         showlegend=False))

# Calculate trendline equation and R-squared value
x = merged_data["Population Density"]
y = merged_data["GHG_Emissions"]
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
trendline_eq = f"y = {slope:.2f}x + {intercept:.2f}"
r_squared = f"R² = {r_value ** 2:.2f}"
fig.update_layout(annotations=[dict(x=0.2, y=0.8, xanchor="left", yanchor="top",
                                     text=trendline_eq, showarrow=False),
                                dict(x=0.2, y=0.7, xanchor="left", yanchor="top",
                                     text=r_squared, showarrow=False)])
fig.show()
```
